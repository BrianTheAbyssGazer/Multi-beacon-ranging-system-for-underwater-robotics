```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

t = np.linspace(0, 100, 1000)
sent_buffer = np.zeros(1000)
for i in range(10):
    sent_buffer[i*100]=1
plt.figure(figsize=(16,5))
xticks = np.arange(0, 100, step=10)
yticks = [0,1]
ylabels = ['on','off']

ax = plt.subplot(3,1,1)
pwm = (signal.square(2 * np.pi * 0.1 * t, duty=0.1)+1)/2
plt.plot(t, pwm)
ax.set_xticks(xticks)
ax.set_yticks(yticks,labels=ylabels)
plt.grid()

ax = plt.subplot(3,1,2)
pwm = (signal.square(2 * np.pi * 0.1 *(t-3), duty=0.1)+1)/2
plt.plot(t, pwm)
ax.set_xticks(xticks)
ax.set_yticks(yticks,labels=ylabels)
plt.grid()
plt.xlabel("time (buffer length)")
```
